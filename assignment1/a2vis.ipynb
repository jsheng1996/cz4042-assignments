```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import pandas as pd
```

```python
data = pickle.load( open( "pickles/a2.p", "rb" ) )
```

```python
plt.plot(data['model_4'])
plt.plot(data['model_8'])
plt.plot(data['model_16'])
plt.plot(data['model_32'])
plt.plot(data['model_64'])
plt.title('Cross Validation Accuracies vs Epoch')
plt.ylabel('Cross Validation Accuracy')
plt.xlabel('Epoch')
plt.legend(['batch_size=4', 'batch_size=8', 'batch_size=16', 'batch_size=32', 'batch_size=64'], loc='lower right')
plt.show()
```

```python
x=[4,8,16,32,64]
plt.plot(x, data['timings'], marker='x')
plt.xticks(np.arange(min(x), max(x)+1, 4))
plt.title('Time per Epoch vs Batch Size')
plt.ylabel('Seconds')
plt.xlabel('Batch Size')
plt.show()
```

```python
%%HTML
<style type="text/css">
table.dataframe td, table.dataframe th {
    border: 1px  black solid !important;
  color: black !important;
}
</style>
```

```python
accuracies_over_time = []
accuracies_over_time.append(("4", data['model_4'][-1]/data['timings'][0], data['model_4'][-1], data['timings'][0], pd.to_timedelta(data['timings'][0]*5000, unit='S')))
accuracies_over_time.append(("8", data['model_8'][-1]/data['timings'][1], data['model_8'][-1], data['timings'][1], pd.to_timedelta(data['timings'][1]*5000, unit='S')))
accuracies_over_time.append(("16", data['model_16'][-1]/data['timings'][2], data['model_16'][-1], data['timings'][2], pd.to_timedelta(data['timings'][2]*5000, unit='S')))
accuracies_over_time.append(("32", data['model_32'][-1]/data['timings'][3], data['model_32'][-1], data['timings'][3], pd.to_timedelta(data['timings'][3]*5000, unit='S')))
accuracies_over_time.append(("64", data['model_64'][-1]/data['timings'][4], data['model_64'][-1], data['timings'][4], pd.to_timedelta(data['timings'][4]*5000, unit='S')))
df = pd.DataFrame(accuracies_over_time, columns =["Batch size", "Accuracy divided by time", "Final Epoch Accuracy", "Time Per Epoch (seconds)", "Total Time(5-Fold Cross Validation)"])
df
```
